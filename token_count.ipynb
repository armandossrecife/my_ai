```python
print("Carrega dependencias...")
# Tokenizer neuralmind/bert-base-portuguese-cased
from datasets import load_dataset
from transformers import AutoTokenizer
from tqdm import tqdm
print("Dependencias carregadas com sucesso!")
```

```python
# Carregar dataset e tokenizador
print("Carregando dataset e tokenizador")
#dataset = load_dataset("dominguesm/legal-bert-base-cased-ptbr", split="train")
dataset = load_dataset("fpaulino/portuguese-tweets", split="train[:2000]")
tokenizer = AutoTokenizer.from_pretrained("neuralmind/bert-base-portuguese-cased")
print("Dataset e tokenizador carregados com sucesso!")
```

```python
# Inspecione a estrutura
print(dataset)
print("\nColunas disponíveis:", dataset.column_names)
```

```python
# Função para contar tokens
def count_tokens(batch):
    tokenized = tokenizer(batch["tweet_text"], truncation=False, return_attention_mask=False)
    return {"num_tokens": [len(tokens) for tokens in tokenized["input_ids"]]}

# Processar em lotes (ajuste o batch_size conforme sua RAM)
batch_size = 1000
token_counts = dataset.map(
    count_tokens,
    batched=True,
    batch_size=batch_size,
    remove_columns=dataset.column_names
)

# Calcular total
total_tokens = sum(token_counts["num_tokens"])
print(f"Total de tokens no dataset: {total_tokens}")

# pode usar apenas as primeiras N amostras:
# small_sample = dataset.select(range(1000))  # Primeiras 1000 amostras
```
